## Evaluation metric (Personalization score)

Personalization is a great way to assess if a model recommends many of the same items to different users. It is the dissimilarity (1- cosine similarity) between userâ€™s lists of recommendations. An example will best illustrate how personalization is calculated.

<img src="https://miro.medium.com/max/1050/1*dhRoTvKdph69JDoXdJ1QTg.png" width="500">
Example list of recommended items for 3 different users.
<img src="https://miro.medium.com/max/519/1*FwO0Fn3R4gOPl6Xic2C3jw.png" width="400">
<img src="https://miro.medium.com/max/616/1*v79uazbvN9ioWzpjioaZjw.png" width = "300">
Then, the cosine similarity matrix is calculated across all userâ€™s recommendation vectors.

Finally, the average of the upper triangle of the cosine matrix is calculated. The personalization is 1-the average cosine similarity.

A high personalization score indicates userâ€™s recommendations are different, meaning the model is offering a personalized experience to each user.

Reference:
https://towardsdatascience.com/evaluation-metrics-for-recommender-systems-df56c6611093#:~:text=three%20recommender%20systems.-,Personalization,the%20model%20is%20offering%20a%20personalized%20experience%20to%20each%20user.,-Intra-list%20Similarity

```python
A = np.array(l_total)
```

```python
A
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine
```

```python
#cosine matrix
dist_out = 1-pairwise_distances(A, metric="cosine")
dist_out
```

```python
i, j = 0, 0;
upper_sum = 0
    
count = 0 
    
for i in range(len(dist_out)):
    for j in range(len(dist_out[0])):
        if (i <= j):
            upper_sum += dist_out[i][j]
            count+=1
print(upper_sum/count)
```

```python
personalization = 1-(upper_sum/count)
personalization
```

```python
print("There for the personalization = "+str(personalization))
```

This means that the recommendations of the candidates are personalized upto 87% which is a good sign.
